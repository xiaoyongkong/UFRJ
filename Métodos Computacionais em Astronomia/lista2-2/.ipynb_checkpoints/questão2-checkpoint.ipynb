```python
import numpy as np
from numpy.lib import scimath as sm

#Arrays da questao 1
array_original = np.loadtxt("array.dat")
novo_array = array_original[:,2]
outro_array = array_original[4]
colunas_pares = array_original[:,1::2]
linhas_impares = array_original[0::2]
multiplos_5 = array_original[array_original%5==0];
maior_5 = array_original[array_original > 0]
transposta = np.transpose(array_original)
achatado = array_original.flatten()
reshape = np.reshape(array_original, (12,10))

loop = [array_original, novo_array, outro_array, colunas_pares, colunas_pares, linhas_impares, multiplos_5, maior_5, transposta, achatado, reshape]
soma_acumulada = 0
```

```python
# Letra a
# Calcule a média, desvio padrão, quadrado, raiz quadrada, seno, soma (dos itens de cada array) e 
# soma acumulada (dos itens de cada array) dos arrays gerados na questão 1.

for word in loop:
    print("########################################################################################\n")
    print("Média dos valores ", np.mean(word))
    print("\nDesvio Padrão dos valores ", np.std(word)) 
    print("\nQuadrado dos valores ", word**2)
    print("\nRaiz Quadrada dos valores ", sm.sqrt(word))
    print("\nSeno dos valores ", np.sin(word))
    print("\nSoma dos valores ", np.sum(word))
    print("\n#######################################################################################\n")
    soma_acumulada += np.sum(word)
print("\nSoma Acumulada dos valores ", soma_acumulada)
    
    
    
```
