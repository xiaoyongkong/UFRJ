```python
# Usando o arquivo “orbita_de_um_asteroide.dat” vamos verificar se a orbita traçada pode oferecer algum risco a
# Terra em algum momento, isto é, se ela intercepta o círculo de 1UA com relação ao Sol no futuro 
# (não entraremos em detalhes da posição da órbita da terra). O asteroide foi recém descoberto e 
# por isso ainda não foi observada a órbita completa, apenas um pequeno trecho da mesma. No arquivo 
# as colunas são theta (em graus), raio em UA e erro do raio em UA, onde o raio é medido com relação ao Sol*.

orbita = np.loadtxt("./orbita_de_um_asteroide.dat")
```

```python
# Letra A
# Faça um polar plot dos pontos observados.

grau = orbita[:,0]
radiano = orbita[:,1]
erro = orbita[:,2]

plt.polar(grau*np.pi/180,radiano,'go')
```

```python
# Letra B
# Ajuste a elipse utilizando o método de minimização do χ (veja mais sobre o método nos comentários 
# e nessa referência) para encontrar os semieixos a e b. Se precisar relembre as propriedades da elipse 
# nessa referência.

a1 = 50
a2 = 20
a3 = 0
b1 = 40
b2 = 10
b3 = 0
while (2*abs((a1-a2)/(a1+a2))> 0.0005) and (2*abs((b1-b2)/(b1+b2))> 0.0005):
    valores_aprox = []
    valores_a = np.arange(a2,a1,(a1-a2)/10)
    valores_b = np.arange(b2,b1,(b1-b2)/10)
    for i in valores_a:
        for j in valores_b:
            e = cm.sqrt(i**2-j**2)/i
            aprox = sum(((radiano - i*(1-e**2)/(1+e*np.cos(grau*np.pi/180)))/erro)**2)
            valores_aprox.append(aprox)
    for i in range(2):
        for j in range(2):
            a1 = valores_a[i+1]
            a2 = valores_a[i-1]
            a3 = valores_a[i]

            b1 = valores_b[j+1]
            b2 = valores_b[j-1]
            b3 = valores_b[j]               
    print(a1, " ", a2, " ", a3, " ", b1, " ", b2, " ", b3, "\n")

```

```python
# Letra C
# Faça um plot da elipse ajustada juntos aos pontos.

angulo = np.arange(0,2*np.pi,np.pi/180)
plt.polar(angulo,a3*(1-e(a3,b3)**2)/(1+e(a3,b3)*np.cos(angulo)))
plt.polar(grau*np.pi/180,radiano,'go')
plt.show()
```

```python
# Letra D
# Após algumas considerações sobre o método utilizado na obtenção dos dados foi considerada a hipótese de que
# o erro poderia estar subestimado, de forma que os semieixos a e b ajustados para a elipse podem 
# ter +/-15% de erro. Plot as duas elipses com mais e menos essa margem nos valores de a e b (faça em 
# uma cor mais clara que a elipse central), preencha o intervalo entre as elipses interna e extercom uma 
# cor com transparência.

a3_min = 0.65*a3
a3_max = 1.25*a3
b3_min = 0.65*b3
b3_max = 1.25*b3
plt.figure(figsize=(7,7))
plt.polar(angulo,a3*(1-e(a3,b3)**2)/(1+e(a3,b3)*np.cos(angulo)),color='green')
plt.polar(angulo,a3_min*(1-e(a3_min,b3_min)**2)/(1+e(a3_min,b3_min)*np.cos(angulo)),color='yellow')
plt.polar(angulo,a3_max*(1-e(a3_max,b3_max)**2)/(1+e(a3_max,b3_max)*np.cos(angulo)),color='red')
plt.fill_between(angulo,a3_min*(1-e(a3_min,b3_min)**2)/(1+e(a3_min,b3_min)*np.cos(angulo)), a3_max*(1-e(a3_max,b3_max)**2)/(1+e(a3_max,b3_max)*np.cos(angulo)),color='blue',alpha=0.2)
plt.show()

```
