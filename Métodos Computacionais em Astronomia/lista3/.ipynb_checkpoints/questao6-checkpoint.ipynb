```python
# Our function will be here
def pontos(arquivo):
    tempo = arquivo[:,0]
    brilho = arquivo[:,1]
    erro = arquivo[:,2]
    return tempo, brilho, erro
    

def calcula_w(arquivo, maior, menor):
    x=arquivo[:,0]
    y=arquivo[:,1]
    e=arquivo[:,2]
    w1 = maior
    w2 = menor
    w3 = 0
    while (2*abs((w1-w2)/(w1+w2))> 0.0005):
        valores_aprox = []
        valores_w = np.arange(w2,w1,(abs(w1-w2))/10)
        for i in valores_w:
              aprox = sum(((y-np.sin(i*x))/e)**2)
              valores_aprox.append(aprox)
        for i in range(len(valores_aprox)):
              if valores_aprox[i] == min(valores_aprox):
                w1 = valores_w[i+1]  
                w2 = valores_w[i-1]
                w3 = valores_w[i]
    return w1, w2, w3

def plotar(arquivo):
    ponto = pontos(arquivo)
    plt.plot(ponto[0], ponto[1])
    plt.show()
    
def subplot(tempo, brilho, erro, time, subpl, calculo):
    plt.plot(time, np.sin(calculo*time), color='red')
    plt.plot(tempo, brilho,'go')
    plt.errorbar(tempo, brilho, yerr=erro)
    plt.grid(which='major',axis='x')
    plt.title('Binária eclipsante')
    plt.xlabel('Tempo(d)')
    plt.ylabel('Brilho')
    sp1.xaxis.set_major_locator(MultipleLocator(2*np.pi/calculo))

def fase(tempo, calculo):
    plot = plt.plot(tempo,np.sin((w_bin1*t1)+t))
    plt.show()
    return plot

def fase(t):
    plot = plt.plot(pontos_1[0],np.sin((w_2[2]*pontos_1[0])+t))
    plt.show()
    return plot

def f(x):
    return x

def g(x, y):
    return (x, y)

    
```

```python
# Letra a
# Usando os arquivos “binaria_eclipsante_1.dat” e “binaria_eclipsante_2.dat” vamos medir o período de dois
# pares de binárias eclipsantes perfeitamente alinhadas e de mesmo brilho (situação idealizada). Nos arquivos as
# colunas são tempo em dias, brilho relativo e erro do brilho relativo, sendo o brilho relativo 0 o brilho relativo
# médio (renormalizado para 0), -1 quando apenas uma estrela é visível e 1 quando ambas estrelas estão visíveis
# (lado a lado). Veja e ste pequeno vídeo ilustrando um sistema binário eclipsante. Em ambos os casos, utilizando o
# 2 método de minimização do χ * (veja mais sobre o método nos comentários), um método um pouco mais sofis-
# ticado que o MMQ, e que é utilizado comumente na astronomia (em artigos). Tente ajustar uma curva do tipo
# sen ( ω t) para encontrar o período do eclipse, onde ω =2 π /T , sendo T o período em dias.

arquivo_1 = np.loadtxt("./binaria_eclipsante_1.dat")
arquivo_2 = np.loadtxt("./binaria_eclipsante_2.dat")
pontos_1 = pontos(arquivo_1)
pontos_2 = pontos(arquivo_2)

print(2*np.pi/5)
print("\n", 2*np.pi/3)
print("\n", 2*np.pi/400)
print("\n", 2*np.pi/200)

w_1 = calcula_w(arquivo_1,2.1,1.2)
print(w_1)
print(2*np.pi/w_1[2])
plt.plot(pontos_1[0], np.sin(w_1[2]*pontos_1[0]))
plt.show()

w_2 = calcula_w(arquivo_2,0.030,0.015)
print(w_2)
print(2*np.pi/w_2[2])
plt.plot(pontos_2[0], np.sin(w_2[2]*pontos_2[0]))
plt.show()
```
