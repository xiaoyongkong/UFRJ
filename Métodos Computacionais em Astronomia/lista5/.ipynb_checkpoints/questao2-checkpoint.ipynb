```python
# Os imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
def proporcao(array, p):
    return np.array([np.quantile(array,(1-p)/2),np.quantile(array,1-(1-p)/2)])

def sortear(x):
    return [np.array([(sort[s][a][sort[s][a] == x ].size/sort[s][a].size) * 100 for a in range(1000)]) for s in range(3)]
```

```python
'''

Vamos simular uma eleição, isto é, um problema de cálculo de proporções utilizando Bootstrap. Gere uma 
população de 100 milhões de votos com 40% dos votos válidos para o candidato A, 31% para o B e 29% para o C. 
Você pode fazer isso com letras ou números (repetindo o número de vezes necessário cada candidato e juntando 
em um único array), e contando o número de repetições para checar a porcentagem. Após isso sorteie amostras 
com 1000, 10000 e 100000 pessoas, as quais você utilizará para estimar o resultado da população total (assim 
como os institutos de estatística). Utilizando a técnica de Bootstrap e considerando um intervalo de confiança 
de 95%, estime os possíveis resultados da eleição de acordo com o tamanho da amostra original e faça um plot 
mostrando as variações do intervalo de confiança e comparando o desempenho dos candidatos (porcentagem e 
margem de erro para confiança de 95%). Feito isso, sorteie agora 1000 amostras da população original (100 
milhões), com os 3 diferentes tamanhos, encontre a distribuição dos resultados (das proporções de votos) e 
verifique se os intervalos de confiança obtidos pelo método de Bootstrap são razoáveis. Verifique também em 
quantos porcento das distribuições sorteadas o candidato C iria para o 2o turno no lugar do B, esta seria a 
probabilidade de a amostra sorteada para a aferição indicar o candidato errado passando para o segundo turno, 
desconsiderando a margem de erro é claro.

'''

populacao = 100000000
tamanho = [1000, 10000, 100000]
cor = ['blue','cyan','magenta']
candidatoA = int(populacao * 0.40)
candidatoB = int(populacao * 0.31)
candidatoC = int(populacao * 0.29)
vA = np.repeat('A', candidatoA)
vB = np.repeat('B', candidatoB)
vC = np.repeat('C', candidatoC)
vP = np.append(vA, vB)
vT = np.append(vP, vC)
np.random.shuffle(vT)
amostras = [vT[np.random.randint(vT.size,size=i)] for i in tamanho]
distribuicao = [[np.random.choice(amostras[s], size = amostras[s].size, replace = True) for i in range(1000)] for s in range(3)]
proporcaoA = [np.array([(distribuicao[s][i][distribuicao[s][i] == 'A'].size/tamanho[s]) * 100 for i in range(1000)]) for s in range(3)]
proporcaoB = [np.array([(distribuicao[s][i][distribuicao[s][i] == 'B'].size/tamanho[s]) * 100 for i in range(1000)]) for s in range(3)]
proporcaoC = [np.array([(distribuicao[s][i][distribuicao[s][i] == 'C'].size/tamanho[s]) * 100 for i in range(1000)]) for s in range(3)]
errA = [proporcao(proporcaoA[s], 0.95) for s in range(3)]
errB = [proporcao(proporcaoB[s], 0.95) for s in range(3)] 
errC = [proporcao(proporcaoC[s], 0.95) for s in range(3)]

for s in range(3):
    plt.hist(proporcaoA[s], color = cor[s])
    plt.axvline(x = errA[s][0], color = cor[s], ls = '--')
    plt.axvline(x = errA[s][1], color = cor[s], ls = '--')

print('===================== A ========================')
plt.show()

for s in range(3):
    plt.hist(proporcaoB[s], color = cor[s])
    plt.axvline(x = errB[s][0], color = cor[s], ls = '--')
    plt.axvline(x = errB[s][1], color = cor[s], ls = '--')

print('===================== B ========================')
plt.show()

for s in range(3):
    plt.hist(proporcaoC[s], color = cor[s])
    plt.axvline(x = errC[s][0], color = cor[s], ls = '--')
    plt.axvline(x = errC[s][1], color = cor[s], ls = '--')
    
print('===================== C ========================')
plt.show()

for s in range(3):
    print(errA[s])
    print(errB[s])
    print(errC[s])
```
